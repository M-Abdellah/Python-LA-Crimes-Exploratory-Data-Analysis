Q1 Answer : Which hour has the highest frequency of crimes?

```python
# Extract hour

crimes['HOUR'] = crimes['TIME OCC'].astype(int) // 100

## Q1 Answer : 

crimes['HOUR'].value_counts().sort_values(ascending=False).head(1)

peak_crime_hour = 12
```

Q2 Answer : Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)?

```python
# Time-Masks

time_mask = [10,11,0,1,2,3]

night_crimes = crimes[crimes['HOUR'].isin(time_mask)]

night_crimes['AREA NAME'].value_counts().sort_values(ascending=False).head(1)

peak_night_crime_location = 'Central'
```

Q3 Answer : Identify the number of crimes committed against victims of different age groups ?

```python
# Create Age buckets

buckets = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

conditions = [
    (crimes['Vict Age'] >= 0) & (crimes['Vict Age'] <= 17),
    (crimes['Vict Age'] >= 18) & (crimes['Vict Age'] <= 25),
    (crimes['Vict Age'] >= 26) & (crimes['Vict Age'] <= 34),
    (crimes['Vict Age'] >= 35) & (crimes['Vict Age'] <= 44),
    (crimes['Vict Age'] >= 45) & (crimes['Vict Age'] <= 54),
    (crimes['Vict Age'] >= 55) & (crimes['Vict Age'] <= 64),
    (crimes['Vict Age'] >= 65)
]

crimes['Age_Bucket'] = np.select(conditions, buckets, default='Unknown')

victim_ages= crimes['Age_Bucket'].value_counts()
```
